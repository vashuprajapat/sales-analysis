<h2>EXPLORATORY DATA ANALYSIS

<h2>What overall sales trend

```python
df['Date'].min()
```

```python
df['Date'].max()
```

```python
#Getting month and year from dataset
df['month_year'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m')
df['date']=pd.to_datetime(df['Date']).dt.strftime('%m-%d-%Y')
```

```python
#Groupping day wise and month wise total sale
df_month_wise_trend = df.groupby('month_year').sum().reset_index()
# date are in formate mm-dd-yyyy
df_day_wise_trend =df.groupby('date').sum().reset_index()
```

```python
df_day_wise_trend
```

```python
df_month_wise_trend
```

<h3>What overall total sales day wise and month wise

```python
#setting figure size
plt.figure(figsize=(15,6))
plt.title('Day wise total sale',size=20, color='black')
plt.xlabel("Date (DD-MM-YYYY)",color='blue',size=10)
plt.ylabel("Sales",color='blue',size=12)
plt.plot(pd.to_datetime(df_day_wise_trend['date']).dt.strftime('%d-%m-%Y'),df_day_wise_trend['Total'],color='#b80045')
plt.xticks(rotation='vertical',size=9)
plt.show()
```

```python
#setting figure size
plt.figure(figsize=(8,4))
plt.title('Month wise total sale',size=20, color='black')
plt.xlabel("Month Year (MM-YYYY)",color='blue',size=10)
plt.ylabel("Sales",color='blue',size=12)
plt.plot(pd.to_datetime(df_month_wise_trend['month_year']).dt.strftime('%m-%Y'),df_month_wise_trend['Total'],color='#b80045')
plt.xticks(rotation='horizontal',size=9)
plt.show()
```

<h3>What overall profit day wise and month wise

```python
#setting figure size
plt.figure(figsize=(15,6))
plt.title('Day wise total gross income',size=20, color='black')
plt.xlabel("Date (DD-MM-YYYY)",color='blue',size=10)
plt.ylabel("Gross income (profit)",color='blue',size=12)
plt.plot(pd.to_datetime(df_day_wise_trend['date']).dt.strftime('%d-%m-%Y'),df_day_wise_trend['gross income'],color='#b80045')
plt.xticks(rotation='vertical',size=9)
plt.show()
```

```python
#setting figure size
plt.figure(figsize=(15,6))
plt.title('Month wise total gross income',size=20, color='black')
plt.xlabel("Date (MM-YYYY)",color='blue',size=10)
plt.ylabel("Gross income (profit)",color='blue',size=12)
plt.bar(pd.to_datetime(df_month_wise_trend['month_year']).dt.strftime('%m-%Y'),df_month_wise_trend['gross income'],color='blue')
plt.xticks(rotation='horizontal',size=9)
plt.show()
```

<h3>What month wise day wise total Quantity sales

```python
#setting figure size
plt.figure(figsize=(15,6))
plt.title('Day wise total Quantity sale',size=20, color='black')
plt.xlabel("Date (DD-MM-YYYY)",color='blue',size=10)
plt.ylabel("Quantity",color='blue',size=12)
plt.bar(pd.to_datetime(df_day_wise_trend['date']).dt.strftime('%d-%m-%Y'),df_day_wise_trend['Quantity'],color='#b80045')
plt.xticks(rotation='vertical',size=9)
plt.show()
```

```python
#setting figure size
plt.figure(figsize=(10,5))
plt.title('Month wise total Quantity sale',size=20, color='black')
plt.xlabel("Date (MM-YYYY)",color='blue',size=10)
plt.ylabel("Quantity",color='blue',size=12)
plt.bar(pd.to_datetime(df_month_wise_trend['month_year']).dt.strftime('%m-%Y'),df_month_wise_trend['Quantity'],color='blue')
plt.xticks(rotation='horizontal',size=9)
plt.show()
```
